```python
import nbformat

# Lista REAL de tus archivos (como en tu imagen)
files = [
    "P2-2-clase02Neurona de McCulloch y Pitts.ipynb",
    "P2-2-Clase2neurona_mcculloch_pitts.ipynb"
]

# Crear notebook final
merged = nbformat.v4.new_notebook()

# Unir todos
for file in files:
    with open(file, encoding="utf-8") as f:
        nb = nbformat.read(f, as_version=4)
        merged.cells.extend(nb.cells)

# Guardar resultado
nbformat.write(merged, "Parcial2-Clase2.ipynb")

print("✅ Listo: Clase4.ipynb creado")
```
