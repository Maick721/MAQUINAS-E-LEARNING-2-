# Normalización de datos con MinMaxScaler
Ejemplo práctico usando sklearn.

```python
from sklearn import preprocessing
import numpy as np
```

```python
# Carga de datos.
X_train = np.array([[ 1., -4.,  5.],
                    [ 5.,  0.,  2.],
                    [ 0.,  6., -3.]])

print(X_train)
```

```python
# Limpieza de datos: normalización.
normalizer = preprocessing.MinMaxScaler()
X_train_norm = normalizer.fit_transform(X_train)
```

```python
# Salida de resultados.
print(X_train_norm)
```

```python
# Verificación: fit + transform por separado
X_train_2 = np.copy(X_train)

f = normalizer.fit(X_train)
X_train_norm_2 = f.transform(X_train_2)

print(np.array_equiv(X_train_norm, X_train_norm_2))
```

```python
# Normalizar un nuevo conjunto de datos de test
X_test = np.array([[ 1., -3.,  2.],
                   [ 3.,  0.,  0.],
                   [ 0.,  4., -1.]])

# IMPORTANTE: solo transform (no fit)
X_test_norm = normalizer.transform(X_test)

print(X_test_norm)
```
