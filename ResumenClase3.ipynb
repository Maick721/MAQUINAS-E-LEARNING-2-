## 1. Estadísticos unidimensionales

### 1.1. Conceptos básicos

En esta sección se introducen los conceptos fundamentales de la estadística. La población es el conjunto total de elementos que se desea estudiar, mientras que la muestra es un subconjunto de la población que debe ser representativo para poder generalizar los resultados. Para lograr esa representatividad se aplica el muestreo aleatorio. La inferencia estadística es el proceso mediante el cual se obtienen conclusiones sobre la población a partir del análisis de la muestra, siempre con un margen de error conocido. Se presentaron ejemplos de muestreos erróneos, como encuestas electorales donde la muestra no es balanceada en género, edad o ubicación geográfica, lo que lleva a conclusiones incorrectas.

### 1.2. Variables

Las variables estadísticas son las cualidades o características que se miden en los individuos de una población. Se clasifican en varios tipos. Las variables ordinales son aquellas que representan cualidades no numéricas pero que siguen un orden, como los niveles educativos. Las variables nominales también son cualidades no numéricas pero no tienen un criterio de orden, como el color de ojos. Las variables cuantitativas se dividen en discretas, que solo pueden tomar un número finito de valores contables como la cantidad de hijos, y continuas, que pueden asumir cualquier valor dentro de un rango continuo como la altura o el peso.

### 1.3. Análisis de variables cualitativas

Para las variables cualitativas, ya sean nominales u ordinales, la representación gráfica más común es el diagrama de sectores, también conocido como gráfico circular. En este tipo de gráfico, cada categoría ocupa una porción del círculo proporcional a su frecuencia, lo que permite visualizar de manera rápida la distribución de los datos.

### 1.4. Análisis de variables cuantitativas

En el análisis de variables cuantitativas se utilizan diferentes medidas para describir el comportamiento de los datos. Las medidas de tendencia central incluyen la media, que es el promedio aritmético de todos los valores; la mediana, que es el valor central cuando los datos están ordenados; y la moda, que es el valor que más se repite. Las medidas de dispersión son la varianza y la desviación típica, que indican qué tan dispersos están los datos con respecto a la media. Las medidas de distribución incluyen el coeficiente de asimetría, que mide si los datos están sesgados hacia la izquierda o hacia la derecha, y la curtosis, que mide el grado de apuntamiento o achatamiento de la distribución en comparación con una distribución normal.

## 2. Características bidimensionales

### 2.1. Introducción a los datos 2D

En esta parte se explica que una imagen digital es una función bidimensional f(x, y), donde x e y representan las coordenadas espaciales de un píxel y f representa su nivel de intensidad. Dependiendo del tipo de imagen, esta intensidad puede ser un solo valor en escala de grises o varios valores para representar color. Los espacios de color más comunes son RGB, donde cada píxel tiene componentes rojo, verde y azul, y otros como HSV o Lab que separan la información de color de la luminosidad.

### 2.2. Procesado de imagen

El procesado de imagen incluye varias transformaciones para modificar o mejorar las imágenes. Las transformaciones de intensidad permiten ajustar el brillo, el contraste y aplicar operaciones como la ecualización del histograma para mejorar la distribución de los niveles de gris. También se pueden cambiar los espacios de color, por ejemplo convirtiendo una imagen RGB a escala de grises, HSV o YCrCb, lo cual es útil para diferentes aplicaciones como segmentación o detección de objetos.

Las transformaciones geométricas modifican la posición de los píxeles en la imagen. La traslación desplaza la imagen en las direcciones x e y según una magnitud determinada. El cropping consiste en recortar una región específica de la imagen, llamada región de interés, para eliminar información que no es relevante o que introduce ruido. La rotación aplica una matriz de transformación con un ángulo determinado, y se puede combinar con un factor de escala y un centro de rotación personalizado. La transformación de perspectiva permite cambiar el punto de vista de la imagen, simulando una rotación en el espacio 3D.

### 2.3. Segmentación de imágenes

La segmentación es el proceso de dividir una imagen en regiones o segmentos para facilitar su análisis. Existen dos tipos principales. La segmentación semántica asigna una etiqueta a cada píxel según la clase a la que pertenece, como por ejemplo separar a las personas del fondo. La segmentación de instancia es más compleja y no solo clasifica los píxeles por clase, sino que además distingue objetos individuales dentro de una misma clase, como identificar a cada persona en un grupo.

Las técnicas de segmentación incluyen la umbralización, que separa los objetos del fondo a partir de un valor de intensidad. El método de Otsu es un procedimiento no paramétrico que encuentra automáticamente el umbral óptimo maximizando la varianza entre clases. La detección de bordes identifica los píxeles donde ocurren cambios bruscos de intensidad. El operador Sobel calcula el gradiente en direcciones horizontal y vertical mediante convolución, obteniendo la magnitud y la dirección del gradiente. El algoritmo de Canny mejora este proceso con supresión non-máximum para adelgazar los bordes y umbralización por histéresis para determinar qué píxeles forman realmente parte de un borde.

### 2.4. Descriptores de forma

Los descriptores de forma analizan la estructura geométrica de los objetos en imágenes binarias. El área es simplemente el número de píxeles que conforman la región de interés. La bounding box es el rectángulo más pequeño que puede contener al objeto. El convex hull es el polígono convexo más pequeño que envuelve al objeto. La excentricidad mide la relación entre la distancia de los focos y el eje principal de la elipse que tiene el mismo momento que el objeto, donde cero indica un círculo y uno un segmento de línea. El diámetro equivalente es el diámetro de un círculo que tiene la misma área que la región de interés.

### 2.5. Descriptores de textura

Los descriptores de textura permiten caracterizar patrones de intensidad en la imagen. La matriz de coocurrencia mide la frecuencia con la que un píxel con una intensidad i es adyacente a otro con intensidad j, considerando una distancia y una orientación. A partir de esta matriz se pueden calcular características como contraste, correlación, disimilitud, homogeneidad, energía y segundo momento angular.

**SIFT (Scale-Invariant Feature Transform)** es un algoritmo que detecta puntos de interés invariantes a escala, rotación y cambios de iluminación. Primero construye una pirámide de diferencias de gaussianas para localizar posibles puntos clave, luego filtra aquellos de bajo contraste y los que están sobre bordes usando la matriz Hessiana, después asigna una orientación a cada punto basándose en histogramas de gradientes y finalmente genera un descriptor de 128 características a partir de una ventana de 16x16 dividida en bloques de 4x4 con histogramas de 8 bins cada uno.

**SURF (Speeded-Up Robust Features)** es una versión optimizada de SIFT que es hasta tres veces más rápida. En lugar de usar diferencias de gaussianas, se basa en el determinante de la matriz Hessiana aproximada con filtros de caja. Utiliza imágenes integrales para acelerar los cálculos. La orientación se determina mediante wavelets de Haar y el descriptor se obtiene a partir de subregiones de 5x5 píxeles, generando 64 o 128 características.

**HOG (Histogram of Oriented Gradients)** es un descriptor que se basa en los gradientes de la imagen. Se calcula el gradiente en cada píxel y se agrupa en celdas de 8x8, construyendo un histograma de 9 bins que cubren direcciones entre 0 y 180 grados. Para hacerlo invariante a cambios de iluminación, se normalizan los histogramas en bloques de 16x16 que se desplazan a lo largo de la imagen. En una imagen de 128x64, el descriptor resultante tiene 3780 características.

**LBP (Local Binary Patterns)** es un descriptor que opera a nivel local comparando cada píxel con sus vecinos. Se calculan las diferencias de intensidad entre el píxel central y sus vecinos, se aplica un umbral asignando 1 a las diferencias positivas y 0 a las negativas, y se forma una cadena binaria que luego se convierte en un valor decimal. Para hacerlo invariante a la rotación, se normalizan las cadenas rotando hasta obtener el valor mínimo. Además, se introdujo el concepto de patrones uniformes, donde solo se consideran aquellos que tienen como máximo dos transiciones entre 0 y 1, reduciendo el número total de combinaciones de 256 a solo 10 valores.
