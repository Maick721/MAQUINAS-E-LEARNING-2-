```python
def find_limits_BP(variable):
    Q1 = stats.scoreatpercentile(variable, 25)
    Q3 = stats.scoreatpercentile(variable, 75)
    RIC = Q3 - Q1
    li = Q1 - 1.5 * RIC
    ls = Q3 + 1.5 * RIC
    
    pos_i = np.where(variable < li)[0]
    pos_s = np.where(variable > ls)[0]
    pos_outliers = np.concatenate((pos_i, pos_s))
    
    return pos_outliers
```
