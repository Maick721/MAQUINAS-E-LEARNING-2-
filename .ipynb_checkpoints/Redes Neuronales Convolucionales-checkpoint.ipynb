<h2 style="text-align:center; font-family:'Times New Roman', serif; font-size:30px; margin:20px 0 0;">
  Investigación de las Redes Neuronales Convolucionales
</h2>



Las **redes neuronales** son un subconjunto del *machine learning* y constituyen el corazón de los algoritmos de *deep learning*.  
Están compuestas por capas de nodos:

- **Capa de entrada**  
- **Una o más capas ocultas**  
- **Capa de salida (output)**  

Cada nodo está conectado a otro y posee un **peso** y un **umbral** asociados.  
El funcionamiento es el siguiente:

- Si el **output** de un nodo supera el valor umbral especificado, ese nodo se **activa** y envía datos a la siguiente capa.  
- Si no supera el umbral, **no transmite información** a la siguiente capa de la red.
- Su objetivo es detectar patrones automáticamente como los bordes, formas, textruas y objetivos


# ¿Cómo funcionan las Redes Neuronales Convolucionales?

Las **redes neuronales convolucionales (CNN)** se distinguen de otras redes neuronales por su rendimiento superior con entradas de señal de **imagen, voz o audio**.  

Se componen de tres tipos principales de capas:

- **Capa convolucional**  
- **Capa de agrupación (pooling)**  
- **Capa totalmente conectada (fully connected)**  

### Funcionamiento
- La **capa convolucional** es la primera capa de una CNN.  
- Puede estar seguida de otras capas convolucionales o de agrupación.  
- La **capa final** siempre es la totalmente conectada.  

Con cada capa, la CNN aumenta su complejidad:  
- Las primeras capas identifican características simples como **colores y bordes**.  
- A medida que los datos avanzan, la red reconoce **formas más grandes**.  
- Finalmente, la CNN logra identificar el **objeto esperado** en la imagen.  


## Capa Convolucional
La **capa convolucional** es el bloque de construcción principal de una CNN y donde se realizan la mayoría de los cálculos.  
Componentes principales:
- **Datos de entrada**: imagen en color como matriz 3D (altura, anchura, profundidad → RGB).  
- **Filtro o kernel**: detector de características que se desplaza por la imagen.  
- **Mapa de características**: resultado de aplicar el filtro a la entrada.  

![Diagrama de CNN](https://assets.ibm.com/is/image/ibm/iclh-diagram-convolutional-neural-networks:16x9?fmt=png-alpha&dpr=on%2C1.25&wid=960&hei=540)

### Proceso
1. El filtro (comúnmente 3x3) se aplica sobre un área de la imagen.  
2. Se calcula un **producto escalar** entre píxeles y filtro.  
3. El resultado se guarda en una matriz de salida.  
4. El filtro se desplaza y repite el proceso.  
5. El resultado final es el **mapa de características** (o mapa de activación).  

### Compartición de parámetros
- Los pesos del filtro permanecen fijos mientras se desplaza → **compartición de parámetros**.  
- Durante el entrenamiento, los pesos se ajustan con **retropropagación** y **descenso del gradiente**.  

### Hiperparámetros
1. **Número de filtros** → afecta la profundidad de salida.  
2. **Stride** → número de píxeles que se mueve el kernel; stride mayor → salida más pequeña.  
3. **Zero-padding** → relleno con ceros para ajustar dimensiones.  
   - **Valid padding**: sin relleno.  
   - **Same padding**: salida del mismo tamaño que la entrada.  
   - **Full padding**: salida más grande añadiendo ceros en los bordes.  

### Activación ReLU
Después de cada convolución se aplica **ReLU** para introducir no linealidad en el modelo.  

### Capas adicionales y jerarquía
- Una capa convolucional inicial puede ir seguida de otras.  
- Esto crea una **estructura jerárquica**:  
  - Capas iniciales → patrones simples (bordes, colores).  
  - Capas posteriores → combinaciones más complejas (formas, objetos).  
- Ejemplo: una bicicleta se reconoce como combinación de cuadro, manillar, ruedas, pedales, etc.  

## Capa de Agrupación (Pooling)
La **agrupación** o **submuestreo** reduce la dimensión y el número de parámetros.  
- Similar a la convolución, pero el filtro **no tiene pesos**.  
- Aplica una función de agregación sobre el campo receptivo.  

### Tipos principales
- **Max pooling**: selecciona el valor máximo.  
- **Average pooling**: calcula el promedio.  

### Beneficios
- Reduce complejidad.  
- Mejora eficiencia.  
- Limita el sobreajuste.  

## Capa Totalmente Conectada
- Cada nodo de salida está conectado directamente con todos los nodos de la capa anterior.  
- Realiza la **clasificación** en base a las características extraídas.  
- Funciones de activación:  
  - **ReLU** en capas convolucionales y de agrupación.  
  - **Softmax** en la capa final para generar probabilidades entre 0 y 1.  


# Proceso de Trabajo en una Red Neuronal Convolucional

El proceso de trabajo comienza con el preprocesamiento de los datos, donde las imágenes se transforman en matrices numéricas que es fundamental para normalizar los valores de los píxeles, generalmente escalándolos entre 0 y 1, y redimensionar todas las muestras a un tamaño estándar, esta etapa asegura que la red reciba información homogénea, lo que facilita que los algoritmos de optimización converjan más rápido y evita que variaciones irrelevantes en la iluminación o el tamaño afecten el aprendizaje inicial del modelo.  

Se desarrolla la extracción de características mediante capas de convolución y activación en esta fase, la red aplica filtros matemáticos que se deslizan sobre la imagen para detectar patrones jerárquicos, desde bordes simples hasta texturas y formas complejas donde tras cada convolución se utiliza una función de activación, como ReLU, para introducir no linealidad y permitir que la red aprenda relaciones más complejas. 

Este proceso se complementa con capas de pooling que simplifican la información espacial, haciendo que el sistema sea invariante a pequeñas traslaciones y capaz de reconocer objetos sin importar su posición exacta.  

La estructura técnica culmina en la etapa de clasificación, es donde los mapas de características abstractas se aplanan en un vector unidimensional y este vector alimenta a una serie de capas densamente conectadas que actúan como el motor de decisión final, mediante una función de salida, como Softmax, la red asigna una probabilidad a cada categoría posible, donde durante el entrenamiento, el error entre la predicción y la realidad se propaga hacia atrás para ajustar los pesos de los filtros, perfeccionando la precisión del modelo en cada iteración hasta lograr un reconocimiento confiable.    


# Operación de Convolución

La base matemática de una CNN no es una multiplicación de matrices común, sino una operación de **producto escalar desplazado**.  
Si tenemos una imagen \(I\) y un filtro (kernel) \(K\), el valor de un píxel resultante en la posición \((i,j)\) se calcula como:




El filtro \(K\) se desliza sobre la imagen \(I\), multiplicando sus pesos por los valores de los píxeles correspondientes y sumando los resultados para generar un **mapa de características**.

# Modelos de Activación (Funciones no lineales)

Las funciones de activación determinan si una neurona debe activarse o no, permitiendo que la red aprenda patrones complejos.

### ReLU (Rectified Linear Unit)

Es la más utilizada en las capas intermedias de las CNN por su eficiencia y porque evita el desvanecimiento del gradiente.

 

Convierte todos los valores negativos en cero y mantiene los positivos iguales.  

### Sigmoide

Se usa principalmente en la capa de salida para clasificación binaria (sí/no).


  
Aplasta cualquier valor de entrada a un rango entre 0 y 1.  

### Softmax

Es el estándar para la capa de salida en clasificación de múltiples categorías (ej. perro, gato, ave).
 

  
Entrega una distribución de probabilidad donde la suma de todas las salidas es igual a 1 (100%).

# Capas de Pooling (Agrupamiento)

El **Max-Pooling** es la operación matemática de reducción más común.  
Dentro de una ventana pequeña (ejemplo: 2x2 píxeles), la red conserva el valor máximo y descarta el resto, reduciendo el tamaño del mapa de datos a la mitad.  


     
# Función de Pérdida (Loss Function)

Para que la red aprenda, necesita medir qué tan lejos está de la respuesta correcta. La más común para informes de CNN es la Entropía Cruzada (Categorical Cross-Entropy):



El objetivo del entrenamiento es **minimizar esta función de pérdida**.


### Ejemplo

```python
# Importamos TensorFlow para crear la red neuronal
import tensorflow as tf

# Importamos herramientas de Keras (ya incluidas en TensorFlow)
from tensorflow.keras import datasets, layers, models

# Librería para graficar
import matplotlib.pyplot as plt

# Librería para trabajar con arreglos
import numpy as np
```

```python
# Cargamos el dataset MNIST (imágenes de números escritos a mano)
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

# Mostramos la forma de los datos de entrenamiento
print("Forma de x_train:", x_train.shape)  # (60000, 28, 28)
print("Forma de y_train:", y_train.shape)  # (60000,)

# Mostramos la forma de los datos de prueba
print("Forma de x_test:", x_test.shape)    # (10000, 28, 28)
print("Forma de y_test:", y_test.shape)    # (10000,)
```

```python
# Mostramos la primera imagen del dataset
plt.imshow(x_train[0], cmap="gray")  # Mostramos en escala de grises

# Mostramos la etiqueta real del número
plt.title(f"Etiqueta: {y_train[0]}")

# Quitamos los ejes para mejor visualización
plt.axis("off")

# Mostramos la imagen
plt.show()
```

```python
# Convertimos los valores de píxeles de 0-255 a 0-1
# Esto ayuda a que el modelo aprenda mejor
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# Las CNN necesitan una dimensión extra (canal)
# Aquí agregamos un canal porque las imágenes son en blanco y negro (1 canal)

x_train = x_train[..., np.newaxis]
x_test = x_test[..., np.newaxis]

# Mostramos la nueva forma de los datos
print("Nueva forma de x_train:", x_train.shape)  # (60000, 28, 28, 1)
print("Nueva forma de x_test:", x_test.shape)    # (10000, 28, 28, 1)
```

```python
# Creamos un modelo secuencial (capas una tras otra)
model = models.Sequential([

    # Primera capa convolucional
    # 32 filtros de tamaño 3x3
    # ReLU como función de activación
    # input_shape define el tamaño de entrada
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),

    # Capa de pooling (reduce tamaño)
    layers.MaxPooling2D((2, 2)),

    # Segunda capa convolucional
    layers.Conv2D(64, (3, 3), activation='relu'),

    # Segunda capa de pooling
    layers.MaxPooling2D((2, 2)),

    # Convertimos los datos 2D en un vector 1D
    layers.Flatten(),

    # Capa densa con 64 neuronas
    layers.Dense(64, activation='relu'),

    # Capa de salida con 10 neuronas (números del 0 al 9)
    # Softmax convierte a probabilidades
    layers.Dense(10, activation='softmax')
])
```

```python
# Mostramos la estructura del modelo
model.summary()
```

```python
# Configuramos el modelo para el entrenamiento

model.compile(
    optimizer='adam',  # Algoritmo de optimización
    loss='sparse_categorical_crossentropy',  # Función de pérdida
    metrics=['accuracy']  # Métrica de evaluación
)
```

```python
# Entrenamos el modelo con los datos de entrenamiento
# epochs = número de veces que el modelo ve los datos

history = model.fit(
    x_train,
    y_train,
    epochs=5,            # Número de iteraciones
    validation_split=0.1 # 10% de los datos para validación
)
```

```python
# Evaluamos el modelo con los datos de prueba

test_loss, test_acc = model.evaluate(x_test, y_test)

# Mostramos resultados
print("Pérdida en prueba:", test_loss)
print("Precisión en prueba:", test_acc)
```

```python
# Graficamos la precisión durante el entrenamiento

plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')

# Títulos y etiquetas
plt.title('Precisión del modelo')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')

# Mostramos la leyenda
plt.legend()

# Mostramos la gráfica
plt.show()
```

```python
# Hacemos predicciones sobre el conjunto de prueba
predicciones = model.predict(x_test)

# Elegimos una imagen de prueba
indice = 0

# Mostramos la imagen
plt.imshow(x_test[indice].squeeze(), cmap="gray")
plt.title("Imagen de prueba")
plt.axis("off")
plt.show()

# Mostramos resultados
print("Etiqueta real:", y_test[indice])
print("Etiqueta predicha:", np.argmax(predicciones[indice]))
```

```python
# Mostramos varias predicciones para comprobar si el modelo acierta
predicciones = model.predict(x_test)

# Cantidad de imágenes a mostrar
cantidad = 10

# Creamos una figura
plt.figure(figsize=(15, 5))

# Recorremos varias imágenes
for i in range(cantidad):
    plt.subplot(2, 5, i + 1)  # 2 filas, 5 columnas
    plt.imshow(x_test[i].squeeze(), cmap="gray")  # Mostrar imagen
    pred = np.argmax(predicciones[i])  # Clase predicha
    real = y_test[i]  # Clase real
    plt.title(f"Real: {real}\nPred: {pred}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Convertimos las predicciones a clases
y_pred = np.argmax(predicciones, axis=1)

# Contamos cuántos aciertos hubo
aciertos = np.sum(y_pred == y_test)
total = len(y_test)

# Mostramos resultados
print("Aciertos:", aciertos)
print("Total:", total)
print("Porcentaje de acierto:", aciertos / total)
```

```python
# Encontramos los índices donde el modelo se equivoca
errores = np.where(y_pred != y_test)[0]

# Mostramos cuántos errores hubo
print("Cantidad de errores:", len(errores))

# Mostrar algunas imágenes mal clasificadas
plt.figure(figsize=(15, 5))

for i in range(10):
    idx = errores[i]
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[idx].squeeze(), cmap="gray")
    plt.title(f"Real: {y_test[idx]}\nPred: {y_pred[idx]}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Creamos la matriz de confusión
cm = confusion_matrix(y_test, y_pred)

# La mostramos
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.title("Matriz de confusión")
plt.show()
```
